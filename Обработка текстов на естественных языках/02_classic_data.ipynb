2\. Для всех РПД из каталога `pdf` извлеките названия и количество тем курса. Представьте результат в виде `pd.DataFrame`, где в качестве индекса выступает название файла, столбцы называются "Тема 1", "Тема 2" и т.д., а в ячейках содержится название темы (без номера).

```python
import os

file_names = os.listdir("./pdf")
file_names
```

```python
topics_all_files = []
for fname in file_names[:-1]:
  reader = PdfReader(f"pdf/{fname}")
  n_pages = len(reader.pages)
  full_text = ''
  for i in range(1,n_pages):
    full_text += reader.pages[i].extract_text()

  id_l = full_text.rfind('5.1. Содержание дисциплины') + len('5.1. Содержание дисциплины')
  id_r = full_text.rfind('5.2. Учебно')
  full_text = full_text[id_l:id_r]

  topic_desc = re.split(r"Тема \d+[.]?", full_text)[1:]
  list_topics = []
  for i in topic_desc:
    par = re.split(r'\n[А-ЯЁ]{1}', i)
    list_topics.append(par[0])
  topics_all_files.append(list_topics)
```

```python
reader = PdfReader(f"pdf/{file_names[-1]}")
n_pages = len(reader.pages)
full_text = ''
for i in range(1,n_pages):
  full_text += reader.pages[i].extract_text()

id_l = full_text.rfind('5.1. Содержание дисциплины') + len('5.1. Содержание дисциплины')
id_r = full_text.rfind('5.2. Учебно')
full_text = full_text[id_l:id_r]

topic_desc = re.split(r"Тема:", full_text)[1:]
list_topics = []
for i in topic_desc:
  par = re.split(r'\n\d{1}', i)
  list_topics.append(par[0])
topics_all_files.append(list_topics)
```

```python
import pandas as pd
```

```python
mx = 0
for i in topics_all_files:
  if len(i)>mx:
    mx = len(i)

cols = [f'Тема {i+1}' for i in range(mx)]
pd.DataFrame(topics_all_files, index=file_names, columns=cols)
```

3\. Считайте файл `pdf/rpd_tobd_22.pdf`. Извлеките из раздела 5.3 таблицу с содержанием семинаров и представьте в виде `pd.DataFrame`. Выведите на экран количество строк в полученной таблице и саму таблицу.

```python
dfs = tabula.read_pdf("pdf/rpd_tobd_22.pdf", pages=[12,13])
df1 = dfs[0]
df1
```

```python
df2 = pd.concat([pd.DataFrame(dfs[1].columns).T, dfs[1].T.reset_index(drop=True).T], axis=0).fillna(' ').reset_index(drop=True)
df2[0] = df2[0]+' '+df2[1]
df2 = df2.drop([1,2,4],axis=1)
df2['mask'] = df2[3].apply(lambda x: '[' in x)
df2['group'] = df2['mask'].cumsum() - df2['mask']
df2
```

```python
df2_stack = []
for i in df2.columns[:-2]:
  df2_stack.append(pd.DataFrame(df2[[i,'group']].groupby('group').apply(lambda x: ' '.join(list(x[i])))))
df2_new = pd.concat(df2_stack,axis=1)
df2_new.columns = df1.columns
df2_new
```

```python
df3_full = pd.concat([df1,df2_new])
df3_full
```

4\. Считайте файл `docx/metodichka_azur_poluitog.docx`. На основе содержимого файла создайте таблицу `pd.DataFrame` со следующими столбцами: "номер упражнение", "текст упражнение", "кол-во рисунков в тексте". Каждый абзац в тексте упражнения должен начинаться с символа "\t". Выведите полученную таблицу на экран.

```python
!pip install docx2python
```

```python
from docx2python import docx2python
```

```python
doc_result = docx2python('metodichka_azur_poluitog.docx')
```

```python
ex_desc = {}
id = 0

for i in doc_result.body[0][-1][0]:
  num_ex = re.findall(r'Упражнение (\d+)', i)
  if num_ex:
    id = num_ex[0]
    ex_desc[id] = [[],0]
    continue

  if id:
    if i[:3].lower() == 'рис':
      ex_desc[id][1] += 1
    elif i[:4] == '----':
      continue
    elif i:
      ex_desc[id][0].append('\t'+i)

  if i=='Заключение':
    break
```

```python
print(''.join(ex_desc['1'][0]))
```

```python
for i in ex_desc:
  ex_desc[i][0] = ''.join(ex_desc[i][0])
df4 = pd.DataFrame(ex_desc, index=['текст упражнение', 'кол-во рисунков в тексте']).T.reset_index(names='номер упражнения')
df4
```

5\. В файлах каталога `imgs` лежат 2 типа сканов: уведомления и пояснительные записки. Создайте `pd.DataFrame` с двумя столбцами: название файла и тип документа. Тип документа определите, воспользовавшись средством для распознавания текста и найдя ключевые слова.

```python
import pytesseract
from PIL import Image
```

```python
import os

image_names = os.listdir("./imgs")
image_names
```

```python
doc_type = []
for img_name in image_names:
  img = Image.open(f'./imgs/{img_name}')
  text_from_img = pytesseract.image_to_string(img, lang='rus').lower()
  if 'уведомление' in text_from_img:
    doc_type.append('уведомление')
  elif 'пояснительная записка' in text_from_img:
    doc_type.append('пояснительная записка')
```

```python
df5 = pd.DataFrame([image_names,doc_type], index=['file name', 'doctype']).T
df5
```
