# 3
**Найдите процент учащихся, выполнивших работу ниже среднего.**

```python
print('Процент людей, сдавших ниже среднего: \n')
for sub, df in dfs.items():
    print(f"{sub} - {round(df[df['Балл'] < df['Балл'].mean()].shape[0]/df.shape[0]*100,2)}%")
```

# 4
**Найти процент учащихся не сдавших экзамен.**

```python
print('Процент людей, не сдавших экзамен: \n')
for sub, df in dfs.items():
    print(f"{sub} - {round(df[df['Балл'] < df['Минимальный балл']].shape[0]/df.shape[0]*100,2)}%")
```

# 5
**Постройте круговую диаграмму, показывающую распределение сдавших и не сдавших экзамен.**

```python
print('Круговые диаграммы, показывающие распределение сдавших и не сдавших экзамен: \n')
fig, axes = plt.subplots(4,1, figsize=(15,20))
labels = ['Сдали', 'Не сдали']
titles = ['русский','математика','физика','информатика']
for i, (sub, df) in enumerate(dfs.items()):
    axes[i].pie(pd.Series(df['Балл'] >= df['Минимальный балл']).value_counts(), labels=labels, autopct='%.0f%%')
    axes[i].set_title(titles[i])
```

# 6
**Постройте ядерную оценку плотности распределению баллов за экзамен.**

```python
print('Ядерные оценки плотности распределения баллов за экзамены: \n')
for sub, df in dfs.items():
    sns.kdeplot(df['Балл'], label=sub)
plt.legend();
```

```python
print('Ядерные оценки плотности распределения баллов за экзамены: \n')

c = ['b','orange','g','r']
titles = ['русский','математика','физика','информатика']
for i, (sub, df) in enumerate(dfs.items()):
    plt.figure(figsize=(10,25))
    plt.subplot(4,1,i+1)
    sns.kdeplot(df['Балл'], c=c[i])
    plt.title(sub)
```

# 7
**Найдите процентное соотношение учащихся, сдавших экзамен на «отлично», «хорошо», «удовлетворительно», «неудовлетворительно».**

```python
print('Соотношение учащихся, сдавших экзамен на «отлично», «хорошо», «удовлетворительно», «неудовлетворительно»: \n')
for sub, df in dfs.items():
    fail = len(df[df['Балл']<50])
    passed = len(df[(df['Балл']>=50) & (df['Балл']<70)])
    good = len(df[(df['Балл']>=70) & (df['Балл']<86)])
    excellent = len(df[df['Балл']>=86])
    print(f'{sub} - {excellent}:{good}:{passed}:{fail}')
```

# 10
**Сколько всего заданий с кратким ответом? С развернутым ответом?**

```python
print('Кол-во школ, принимвших участие в экзамене: \n')
for sub, df in dfs.items():
    sh = len(df['Задания с кратким ответом'].iloc[0])
    ln = df['Задания с развёрнутым ответом'].iloc[0].count('(')
    print(f'{sub}:\n кол-во заданий с кратким ответом  - {sh}\n кол-во заданий с развёрнутым  ответом  - {ln}\n')
```

 Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В.**

```python
# Для экзамена по математике
num_tasks = 12 # кол-во заданий с кратким ответом (из предыдущего задания)
print('Процент выполненных и невыполненных заданий: \n')
for i in range(num_tasks):
    print(f'Задание №{i+1}')
    print(df2['Задания с кратким ответом'].dropna().apply(lambda x: 'Правильно' if (x[i] != 0) and (x[i] != '-') else 'Неправильно').value_counts(),'\n')
```

# 13
**Сделайте анализ по двум школам:**
* по всем выполненным заданиям типа В
* по заданиям типа С больше 50%
* по среднему баллу юношей и девушек

```python
df2.head() # сравним школы №136 и №130
```

```python
sch = [136, 130]
```

```python
# сравним кол-во людей, правильно выполнивших все задания типа B
print('Кол-во людей, правильно выполнивших все задания типа B:')
for num in sch:
    print(f'\nШкола №{num}: ', end='')
    print(df2[(df2['№ школы']==num)&(df2['Задания с кратким ответом']=='+'*12)].shape[0])
```

```python
# сравним процент людей, набравших больше 50% по заданиям типа C
num_tasks = 7 # кол-во заданий с развёрнутым ответом (из 10 задания)
print('Процент людей, набравших больше 50% по заданиям типа C: \n')

for num in sch:
    print(f'Школа №{num}')
    ts1 = df2[df2['№ школы']==num]['Задания с развёрнутым ответом'].dropna().apply(lambda x: re.sub(r'\(\d\)', ' ', x).strip().split(' '))
    print(ts1.apply(lambda x: 'Больше 50%' if sum(map(int, x)) > 10 else 'Меньше 50%').value_counts()/ts1.shape[0],'\n')
```

```python
print('Средний балл юношей и девушек:')
for num in sch:
    print(f'\nШкола №{num}')
    print(df2[df2['№ школы']==num].groupby('Пол')['Балл'].mean())
```
